```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
```

```python
# Лабораторная работа №2. Реализация метода главных компонент

# 2. Загрузка набора данных `breast cancer wisconsin dataset`
data = load_breast_cancer().data
target = load_breast_cancer().target

def myPca(data, n):
    # 3. метод главных компонент в функции, принимающей 2 аргумента: 
    # входные данные и искомое число главных компонент N
    dataCentred = []
    vecsArr = []

    # 3.1 нормализация данных
    for x in data:
        dataCentred.append(x - x.mean())

    # 3.2 построение матрицы ковариации
    covmat = np.cov(dataCentred)

    # 3.3 диагонализация матрицы
    _, vecs = np.linalg.eig(covmat)

    # 3.5 оставляем N главных компонент
    for i in range(n):
        v = vecs[:,i]

        # 3.6 модифицируем исходные данные
        dataNew = np.dot(v,dataCentred)
        vecsArr.append(dataNew)
    
    # 3.7 возвращаем данные сниженной размерности
    return vecsArr

dataNew = myPca(data, 2)
print('Our reduced: \n', dataNew)
```

```python
# 4. сравнение с методом в Scikit-learn
pca = PCA(n_components = 2)
XPCAreduced = pca.fit_transform(np.transpose(data))
print('Sklearn reduced: \n', XPCAreduced)
```

```python
dataNew = myPca(data, 3)
print('Our reduced: \n', dataNew)
pca = PCA(n_components = 3)
XPCAreduced = pca.fit_transform(np.transpose(data))
print('Sklearn reduced: \n', XPCAreduced)
```
